# A Recommender System Popularity Based

<p style="font-size: 17px">In that project I will try to build a recommender system based on the popularity of the products. I will use Amazon's dataset.</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("data.csv")
```

```python
df.columns = ['user_id', 'product_id','ratings','timestamp']
```

```python
df.head()
```

```python
df.shape
```

```python
df.sample()
```

```python
df=df.sample(50000)
```

```python
df=df.groupby("product_id").agg({"ratings":[np.size,np.sum,np.mean]})
```

```python
df.head()
```

```python
popular_products=df.sort_values(("ratings","mean"),ascending=False)
```

```python
popular_products.head()
```

```python
grouped_sum=df["ratings"]["sum"].sum()
```

```python
popular_products["percent"]=df["ratings"]["sum"].div(grouped_sum)*100
```

```python
popular_products.head()
```

```python
popular_products=popular_products.sort_values("percent",ascending=False)
```

```python
popular_products
```

```python
popular_products["Rank"]=popular_products["percent"].rank(ascending=False)
```

```python
popular_products.head()
```
